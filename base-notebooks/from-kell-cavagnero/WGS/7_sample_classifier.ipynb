```python
import biom

import pandas as pd

from qiime2 import Artifact

from qiime2.plugins import metadata

import seaborn as sns

import skbio
from skbio.stats.distance import MissingIDError
from skbio import DistanceMatrix

import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import csv
import qiime2 as q2

from qiime2.plugins.sample_classifier.pipelines import classify_samples

from qiime2.plugins.feature_table.methods import filter_samples

from qiime2.plugins.sample_classifier.methods import predict_classification

from qiime2.plugins.sample_classifier.visualizers import confusion_matrix

import os
```

```python
#import table 
q2bt = q2.Artifact.load('tree-filtered-table.qza')

bt = biom.load_table('tree-filtered-table.biom')

#import metadata
q2mf = q2.Metadata.load('filtered-metadata.qza')

mf = q2.Metadata.load('filtered-metadata.qza').to_dataframe()
```

```python
#run classifier and save estimator
accuracy = classify_samples(q2bt, q2mf.get_column('host_subject_id'))
sample_estimator = accuracy.sample_estimator
```

```python
import os
import tempfile
from tqdm.notebook import tqdm

all_dfs = {}
all_confs = {}
#keep dataframes for each preservation method
for (preservation, temperature, duration), duration_df in tqdm(mf.groupby(['sample_preservation_method',
                                                                      'sample_storage_temp_treatment_week_added',
                                                                      'duration_of_storage'])):

            #filter biom table and metadata, run predictor

            # prevent filtering org.
            bt_filter = bt.copy()

            #identify the metadata data to keep (for BL)
            keep = [ind for ind in duration_df.index]
            #indentify the biom table data to keep (match to metadata)
            keep2 = list(set(keep) & set(bt_filter.ids()))

            #filter metadata
            filt_duration_df = duration_df.loc[keep2, : ]

            #filter biom table to match subsetted metadata
            bt_filter = bt_filter.filter(filt_duration_df.index, axis='sample')

            #some features may have been removed that were unique to the samples removed
            #now there may be a lot of 0's in our table, remove them and match tables (unnecessary?)
            keep_feat = bt_filter.ids('observation')[bt_filter.sum(axis='observation') > 0]
            bt_filter = bt_filter.filter(keep_feat, axis='observation')             

            #import into q2 artifacts
            q2mf = q2.Metadata(filt_duration_df)
            q2bt = q2.Artifact.import_data('FeatureTable[Frequency]', bt_filter)

            #run predictor
            predict = predict_classification(q2bt,sample_estimator)

            # save result
            matrix = confusion_matrix(predict.predictions,
                                      q2mf.get_column('host_subject_id'))
            
            with tempfile.TemporaryDirectory() as temp_dir_name:
                out_dir = os.path.join(temp_dir_name, 'vis_out')
                matrix.visualization.export_data(out_dir)
                #tst_ = os.listdir(out_dir)
                df_ = pd.read_csv(os.path.join(temp_dir_name, 'vis_out', 'predictive_accuracy.tsv'),
                                  sep='\t', index_col=0)
            acc_ = df_.loc[['Overall Accuracy',
                            'Baseline Accuracy',
                            'Accuracy Ratio'],
                           ['Overall Accuracy']]
            conf_ = df_.drop(['Overall Accuracy',
                              'Baseline Accuracy',
                              'Accuracy Ratio']).dropna(axis=1)
            all_dfs[(preservation, temperature, duration)] = acc_
            all_confs[(preservation, temperature, duration)] = conf_
            
combined_predictions = pd.concat(all_dfs, axis=0).reset_index()
combined_predictions = combined_predictions.rename({'level_0':'preservation',
                                                  'level_1':'temperature',
                                                  'level_2':'duration',
                                                   'level_3':'stat'}, axis=1)
combined_predictions.head(5)
```

```python
combined_predictions.to_csv('predictions.tsv',sep='\t', quoting=csv.QUOTE_NONE)
```
